```python
import numpy as np
import matplotlib.pyplot as plt

# Load images from .npy file
images = np.load('new_test.npy')
images = images[:,:,:,0]

# Check if there are at least 16 images, if not, adjust the number accordingly
n_images = min(16, images.shape[0])

# Create a 4x4 grid of subplots
fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(10, 10))

# Flatten the array of axes to make indexing easier
axes = axes.flatten()

# Loop through the first 16 images (or less if fewer images are available)
for i in range(n_images):
    # Display each image in a subplot
    ax = axes[i]
    ax.imshow(images[i], cmap='gray' if images[i].ndim == 2 else None)
    ax.axis('off')  # Hide axes

# Fill any remaining subplot (if fewer than 16 images) with white space and remove axes
for j in range(i + 1, 16):
    axes[j].axis('off')

# Adjust spacing between images
plt.tight_layout()
plt.show()
```
